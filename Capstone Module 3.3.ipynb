```python
import numpy as np
import pandas as pd

import requests
from bs4 import BeautifulSoup
```

```python
my_table = requests.get('https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M').text
soup = BeautifulSoup(my_table, 'html.parser')

postcode = []
borough = []
neighborhood = []
```

```python
soup.find('table').find_all('tr')

for row in soup.find('table').find_all('tr'):
    cells = row.find_all('td')
    if(len(cells) > 0):
        postcode.append(cells[0].text)
        borough.append(cells[1].text)
        neighborhood.append(cells[2].text.rstrip('\n'))
```

```python
df = pd.DataFrame({"PostalCode": postcode,
                           "Borough": borough,
                           "Neighborhood": neighborhood})

df.head()
```

```python
df_1 = df[df.Borough != "Not assigned"].reset_index(drop=True)
df_1.head()
```

```python
df_2 = df_1.groupby(["PostalCode", "Borough"], as_index=False).agg(lambda x: ", ".join(x))
df_2.head()
```

```python
for index, row in df_2.iterrows():
    if row["Neighborhood"] == "Not assigned":
        row["Neighborhood"] = row["Borough"]
        
df_3 = df_2
df_3
```

```python
radius = 500
LIMIT = 100

venues = []

for lat, long, post, borough, neighborhood in zip(df_6['Latitude'], df_6['Longitude'], df_6['PostalCode'], df_6['Borough'], df_6['Neighborhood']):
    url = "https://api.foursquare.com/v2/venues/explore?client_id={}&client_secret={}&v={}&ll={},{}&radius={}&limit={}".format(
        CLIENT_ID,
        CLIENT_SECRET,
        VERSION,
        lat,
        long,
        radius, 
        LIMIT)
    
    results = requests.get(url).json()["response"]['groups'][0]['items']
    
    for venue in results:
        venues.append((
            post, 
            borough,
            neighborhood,
            lat, 
            long, 
            venue['venue']['name'], 
            venue['venue']['location']['lat'], 
            venue['venue']['location']['lng'],  
            venue['venue']['categories'][0]['name']))
```

```python
df_7 = pd.DataFrame(venues)

df_7.columns = ['Postal Code', 'Borough', 'Neighborhood', 'Borough Latitude', 'Borough Longitude', 'Venue Name', 'Venue Latitude', 'Venue Longitude', 'Venue Category']

print(df_7.shape)
df_7.head()
```

```python
df_8 = pd.get_dummies(df_7[['Venue Category']], prefix="", prefix_sep="")

df_8['Postal Code'] = df_7['Postal Code'] 
df_8['Borough'] = df_7['Borough'] 
df_8['Neighborhoods'] = df_7['Neighborhood'] 

fixed_columns = list(df_8.columns[-3:]) + list(df_8.columns[:-3])
df_8 = df_8[fixed_columns]

df_9 = df_8.groupby(["Postal Code", "Borough", "Neighborhoods"]).mean().reset_index()
```

```python
venues_top = 10

indicators = ['st', 'nd', 'rd']

area_columns = ['Postal Code', 'Borough', 'Neighborhoods']
freq_columns = []
for ind in np.arange(venues_top):
    try:
        freq_columns.append('{}{} Most Common Venue'.format(ind+1, indicators[ind]))
    except:
        freq_columns.append('{}th Most Common Venue'.format(ind+1))
columns_2 = area_columns + freq_columns


df_10 = pd.DataFrame(columns=columns_2)
df_10['Postal Code'] = df_9['Postal Code']
df_10['Borough'] = df_9['Borough']
df_10['Neighborhoods'] = df_9['Neighborhoods']

for ind in np.arange(df_9.shape[0]):
    row_categories = df_9.iloc[ind, :].iloc[3:]
    row_categories_sorted = row_categories.sort_values(ascending=False)
    df_10.iloc[ind, 3:] = row_categories_sorted.index.values[0:venues_top]

print(df_10.shape)
```

```python
kclusters = 5

Toronto_clustering = df_9.drop(["Postal Code", "Borough", "Neighborhoods"], 1)

kmeans = KMeans(n_clusters=kclusters, random_state=0).fit(Toronto_clustering)

kmeans.labels_[0:10]
```
